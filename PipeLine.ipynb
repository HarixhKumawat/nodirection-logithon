```python
def estimate_air_route_cost(distance_km, weight_kg, fuel_consumption_km, fuel_price_per_unit, operating_costs):
    # Calculate fuel usage for the given distance and weight
    fuel_usage = fuel_consumption_km * distance_km

    # Calculate fuel cost
    fuel_cost = fuel_usage * fuel_price_per_unit

    # Calculate total cost including fuel and operating costs
    total_cost = fuel_cost + operating_costs

    return total_cost

# Example inputs
distance_km = 5000  # Distance in kilometers
weight_kg = 5000    # Weight in kilograms
fuel_consumption_km = 0.1  # Assume 0.1 liters/km for demonstration purposes
fuel_price_per_unit = 2     # Assume $2 per liter for demonstration purposes
operating_costs = 5000      # Assume $5000 per flight for demonstration purposes

# Calculate estimated cost
estimated_cost = estimate_air_route_cost(distance_km, weight_kg, fuel_consumption_km, fuel_price_per_unit, operating_costs)
print(f"Estimated cost for a {distance_km} km air route with {weight_kg} kg weight: ${estimated_cost}")
```

```python
def estimate_road_trip_cost(distance_km, fuel_consumption_km, fuel_price_per_unit, toll_fees, maintenance_costs):
    # Calculate fuel usage for the given distance
    fuel_usage = fuel_consumption_km * distance_km

    # Calculate fuel cost
    fuel_cost = fuel_usage * fuel_price_per_unit

    # Calculate total cost including fuel, toll fees, and maintenance costs
    total_cost = fuel_cost + toll_fees + maintenance_costs

    return total_cost

# Example inputs
distance_km = 500  # Distance in kilometers
fuel_consumption_km = 0.08  # Assume 0.08 liters/km for demonstration purposes
fuel_price_per_unit = 2     # Assume $2 per liter for demonstration purposes
toll_fees = 50              # Assume $50 for toll fees for demonstration purposes
maintenance_costs = 200      # Assume $200 for maintenance costs for demonstration purposes

# Calculate estimated cost
estimated_cost = estimate_road_trip_cost(distance_km, fuel_consumption_km, fuel_price_per_unit, toll_fees, maintenance_costs)
print(f"Estimated cost for a {distance_km} km road trip: ${estimated_cost}")
```
